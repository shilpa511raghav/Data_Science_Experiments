# Seaborn barplot

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

#load dataset from seaborn github instead of creating a dataframe 
tips_data= sns.load_dataset("tips").head(70)
tips_data
```

```python
sns.barplot(x="total_bill",y= "day", data = tips_data, palette= "spring")
```

# ci attribute
#)It is ues to determine the size of confidence interval to draw around estimated values.(black line in between the values shows error part) and its percentile must be in the range [0, 100]

* less will be value of ci small is the interval and vice versa

```python
sns.barplot(x="total_bill",y= "day", data = tips_data, color= "red", ci=5)
```

## Saturation attribute
* It gives us the brighness and darkness level of a color.
* higher the saturation value,brighter is the color and lower the saturation value,darker is the color 

```python
sns.barplot(x= "smoker", y="tip", data = tips_data, color= "red", saturation= 7)
```

```python
sns.barplot(x= "smoker", y="tip", data = tips_data, color= "red", saturation= 0.5)
```

# Order attribute:
.)Order attribute changes the order. for example here order of day i.e in x axis is changes as per specified in the order attribute.

```python
import matplotlib.pyplot as plt
sns.barplot(x= "day", y="total_bill", data = tips_data, palette='spring',order= ['Sun','Thur','Sat','Thur'])
```

```python
# without estimator
sns.barplot(x= "smoker", y="tip", data = tips_data)
```

## Estimator attribute

```python
from numpy import median
sns.barplot(x= "smoker", y="tip", data = tips_data, estimator=median)
```

```python
sns.barplot(x="smoker", y="tip",data= tips_data,hue="sex")
```
