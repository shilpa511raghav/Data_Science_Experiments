# Seaborn


>Seaborn is a data visualization library uses matplotlib

```python
#import various libraries
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
iris= pd.read_csv('D:\\Experiment\\seaborn_basics1)\\Iris_data\\iris.csv')
iris.head()
```

### Joint plot

#) Here we have given sepal_length along x axis and sepal_width along y axix andit is picking the data from iris dataframe 

```python
sns.jointplot(x= 'petal_length',y='sepal_width',data= iris) # by default joint plot has scatter plot in centre and histogram on sides
```

#In  graph petal length and petal width showing positive correlation as pearsonr cofficient is +0.96 means petal width is increasing with petal length 

```python
sns.jointplot(x= 'petal_length',y='petal_width',data= iris, size= 4) # size of plot 
```

1) if we want to draw a regression line on this joint plot then we will use kind attribute and kind must be either 'scatter', 'reg', 'resid', 'kde', or 'hex' 

```python
sns.jointplot(x= 'petal_length',y='petal_width',data= iris, size= 4,kind ='reg')
```

2) now to add colour 

```python
sns.jointplot(x= 'petal_length',y='petal_width',data= iris,color='green', size= 4)
```

3) if we want to convert pattern then use kind attribute and kind must be either 'scatter', 'reg', 'resid', 'kde', or 'hex' 

```python
sns.jointplot(x= 'petal_length',y='petal_width',data= iris, size= 8,kind ='kde')
```

### Stat_func attribute for joint plot

```python
from scipy.stats import spearmanr

sns.jointplot(x = "sepal_length",y = "petal_width",data =iris, color ='red',kind='hex',stat_func = spearmanr)
```

### Ratio attribute for joint plot

Higher the ratio value more smooth is the histogram in joint plot

```python
sns.jointplot(x='sepal_length',y= "petal_width",data= iris,ratio= 20,size =5)
```

```python
sns.jointplot(x='sepal_length',y= "petal_width",data= iris,ratio= 1,size =5)
```

### seaborn.set_style

a)style must be one of white, dark, whitegrid, darkgrid, ticks


```python
import numpy as np
from scipy.stats import kendalltau
import seaborn as sns
sns.set(style="dark")  # style is bacically for histograms plotted on two sides of the graph

rndmstat = np.random.RandomState(1)
x = rndmstat.gamma(16, size=1)  # gamma is a gamma distribution function based on its formula it takes parameters such as size ,shape etc
y = -.5 * x + rndmstat.normal(size=1) #if we increase the gamma(15) then plotvalues increases which are given on axis

sns.jointplot(x, y, kind="scatter", stat_func=kendalltau, color="#b24bb2") #  color code is in hex form u can search it from google
```
