```python
# to give x and y labels and title
from matplotlib import pyplot as plt
x=[2,3,4]
y=[4,6,7]

plt.plot(x,y)
plt.title('distance graph')
plt.xlabel('time in minutes')
plt.ylabel('speed m/s')
plt.show()
```
