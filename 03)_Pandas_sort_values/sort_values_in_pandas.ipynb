```python
import pandas as pd
import numpy
df = pd.DataFrame({
    'alpha' : ['A', 'D', 'B', numpy.nan, 'D', 'A'],
    'c1' : [3, 10, 9, 8, 7, 4],
    'c2': [0, 1, 9, 4, numpy.nan, 3],
})
df[:]
#
```

```python
#Sort by 1st column(i.e= alpha)
df.sort_values(by='alpha')  #here sorting  is based on column 1's alphabets position
```

```python
df.sort_values(by= ['alpha','c1'], ascending=True)
#here we have two columns so in c1 only the position of D in ascending order= True is changed, 
```

```python
#if we want na in first position
df.sort_values(by=['c2'],ascending=False, na_position= 'first')
```

```python
df.sort_values(by=['alpha','c2'],axis=0, na_position='last') # na position is assigned to respectve argument written first
```
