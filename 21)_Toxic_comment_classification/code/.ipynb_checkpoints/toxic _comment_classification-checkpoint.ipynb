```python
# Exract features
#select some data points so that it doesnot take time to load all the dataset
train_df= train_df.head(3000)
def new_features(input):
    train_df= input
    train_df['body_length']= train_df['comment_text'].apply(lambda x: len(x)- x.count(" "))
    train_df['capital_word_count']=train_df.apply(lambda x:sum(1 for each in x['comment_text'].split() if each.isupper()),axis=1)
    train_df['comment_length']= train_df.comment_text.apply(len)
    train_df['single_new_line'] = train_df['comment_text'].apply(lambda comment: comment.count('\n'* 1))
    train_df["double_new_line"] =train_df['comment_text'].apply(lambda x: x.count('\n'*2))
    train_df['capitals']= train_df.comment_text.apply(lambda comment: sum(1 for c in comment if c.isupper()))
    train_df['exclamation']= train_df.comment_text.apply(lambda comment: comment.count('!'))
    train_df['num_punctuation']= train_df['comment_text'].apply(lambda comment: len([c for c in str(comment) if c in string.punctuation]))
    train_df['num_question_marks']= train_df['comment_text'].apply(lambda comment: comment.count("?"))
    train_df['num_unique_words'] = train_df['comment_text'].apply(lambda comment: len(set(w for w in comment.split())))
    train_df['num_stopwords'] = train_df['comment_text'].apply(lambda comment: len([w for w in str(comment).lower().split() if w in eng_stopwords]))
    train_df['caps_vs_length'] = train_df.apply(lambda row: float(row['capitals'])/float(row['comment_length']),axis=1)
    train_df['num_words'] = train_df['comment_text'].apply(lambda comment: len(comment.split()))
    train_df['words_vs_unique'] = train_df.apply(lambda row: float(row['num_unique_words'] )/float(row['num_words']),axis=1)
    #train_df['remove_punc'] = train_df['comment_text'].apply (lambda x: [char for char in x if char not in string.punctuation])
    return train_df
```

```python
import nltk
ps= nltk.PorterStemmer()
def clean_text(text):
    text= train_df['comment_text']
    text= "".join([word.lower() for word in text if word not in string.punctuation])
    tokens= re.split('\W+', text)
    text= [ps.stem(word) for word in tokens if word not in eng_stopwords]
    return text
```
